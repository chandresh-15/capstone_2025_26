```python
# -------- Step 1: Chunk Reader --------
def read_json_in_chunks(path, chunk_size=50000):
    batch = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            batch.append(json.loads(line))
            if len(batch) == chunk_size:
                yield batch
                batch = []
    if batch:
        yield batch

# -------- Step 2: Convert Nested Values --------
def convert_nested(obj):
    if isinstance(obj, (dict, list)):
        return json.dumps(obj)
    return obj

# -------- Step 3: Connect to SQLite --------
conn = sqlite3.connect("yelp.db")
cursor = conn.cursor()
cursor.execute("DROP TABLE IF EXISTS reviews;")
conn.commit()

# -------- Step 4: Process & Insert in Chunks --------
review_path = "../yelp_dataset/yelp_academic_dataset_review.json"

for batch in tqdm(read_json_in_chunks(review_path, chunk_size=50000)):
    df = pd.DataFrame(batch)
    df = df.applymap(convert_nested)
    df.to_sql("reviews", conn, if_exists="append", index=False)

# -------- Step 5: Verify --------
count_df = pd.read_sql_query("SELECT COUNT(*) AS total FROM reviews;", conn)
print(count_df)

# -------- Step 6: Fetch Again --------
sample_review = pd.read_sql_query("SELECT * FROM reviews LIMIT 5;", conn)
sample_review
```
